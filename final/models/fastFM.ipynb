```python
user_mapping = None
business_mapping = None

top_n_categories = 10
all_features = False


# Map user_id and business_id to unique indexes
user_mapping = {user: idx for idx, user in enumerate(phil_reviews_df['user_id'].unique())}
business_mapping = {biz: idx for idx, biz in enumerate(phil_reviews_df['business_id'].unique())}
phil_reviews_df['user_id_enc'] = phil_reviews_df['user_id'].map(user_mapping)
phil_reviews_df['business_id_enc'] = phil_reviews_df['business_id'].map(business_mapping)

# Normalize numerical features
for col in ['review_count', 'useful', 'funny', 'cool']:
    phil_reviews_df[col] = phil_reviews_df[col].fillna(0)  # Replace NaN with 0
    phil_reviews_df[col] = (phil_reviews_df[col] - phil_reviews_df[col].mean()) / phil_reviews_df[col].std()

# Add review length feature
phil_reviews_df['review_length'] = phil_reviews_df['text'].apply(lambda x: len(x.split()) if pd.notnull(x) else 0)
phil_reviews_df['review_length'] = (phil_reviews_df['review_length'] - phil_reviews_df['review_length'].mean()) / phil_reviews_df['review_length'].std()

# If param is true, get the top N unique categories by frequency
if all_features:
    phil_reviews_df['categories'] = phil_reviews_df['categories'].fillna('')  # Handle NaN values
    phil_reviews_df['categories_list'] = phil_reviews_df['categories'].apply(lambda x: [cat.strip() for cat in x.split(',')])
    category_counts = {}
    for cat_list in phil_reviews_df['categories_list']:
        for cat in cat_list:
            category_counts[cat] = category_counts.get(cat, 0) + 1
    top_categories = sorted(category_counts, key=category_counts.get, reverse=True)[:top_n_categories]

    # One-hot encode top N categories
    for category in top_categories:
        phil_reviews_df[f"category_{category}"] = phil_reviews_df['categories_list'].apply(lambda x: 1 if category in x else 0)

    # Create the final feature set
    category_features = [f"category_{cat}" for cat in top_categories]
    features = ['user_id_enc', 'business_id_enc', 'review_count', 'useful', 'funny', 'cool', 'review_length'] + category_features
    phil_reviews_df.drop(columns=['categories_list'], inplace=True)
else:  # A subset of features for testing
    features = ['user_id_enc', 'business_id_enc', 'review_count', 'useful', 'review_length']

target = 'stars'

features, target = phil_reviews_df[features], phil_reviews_df[target]
```

```python
import scipy

# Count the number of unique users and items
nUsers = len(features['user_id_enc'].unique())
nItems = len(features['business_id_enc'].unique())

# Create the sparse matrix
    # One hot encoding of nUsers
    # One hot encoding of nItems
    # 3 additional features for review_count, useful, review_length
X = scipy.sparse.lil_matrix((len(features), nUsers + nItems + 3))
```

```python
# Create offests for indexing into the X sparse matrix
feature_offset_review_count = nUsers + nItems + 0
feature_offset_useful = nUsers + nItems + 1
feature_offset_review_length = nUsers + nItems + 2

# Iterate over all features to populate the sparse matrix
for i in range(len(features)):

    if (i + 1) % 10000 == 0: print(f"Completed: {i+1} / {len(features)}")

    record = features.iloc[i]

    user = int(record['user_id_enc'])
    item = int(record['business_id_enc'])

    X[i,user] = 1 # One-hot encoding of user
    X[i,nUsers + item] = 1 # One-hot encoding of item

    # "review_count" representation
    X[i, feature_offset_review_count] = record['review_count']

    # "useful" representation
    X[i, feature_offset_useful] = record['useful']

    # "review_length" representation
    X[i, feature_offset_review_length] = record['review_length']
```

```python
# Convert the target into a numpy array
import numpy
y = numpy.array(target)
```

```python
from sklearn.model_selection import train_test_split


# Perform the train-test split with a ratio of 95:5
X_train, y_train = X[:90000], y[:90000]
X_test, y_test = X[90000:], y[90000:]

# Further split the training data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
import optuna
from fastFM import als
from sklearn.metrics import mean_squared_error
import numpy as np


def objective(trial):
    # Suggest hyperparameters to tune
    n_iter = trial.suggest_int("n_iter", 500, 1000)
    init_stdev = trial.suggest_float("init_stdev", 0.01, 0.1)
    rank = trial.suggest_int("rank", 2, 20)
    l2_reg_w = trial.suggest_float("l2_reg_w", 0.01, 0.5)
    l2_reg_V = trial.suggest_float("l2_reg_V", 0.01, 0.5)

    # Create and train the model
    fm = als.FMRegression(
        n_iter=n_iter,
        init_stdev=init_stdev,
        rank=rank,
        l2_reg_w=l2_reg_w,
        l2_reg_V=l2_reg_V,
    )
    fm.fit(X_train, y_train)

    # Predict on the validation set
    y_pred = fm.predict(X_val)

    # Replace out of range values with mean on training set
    cnt = 0
    for i in range(len(y_pred)):
        if y_pred[i] < 1 or y_pred[i] > 5:
            cnt += 1
            y_pred[i] = sum([i for i in y_train]) / len([i for i in y_train])

    # Calculate the mean squared error
    mse = mean_squared_error(y_val, y_pred)

    # For ratings outside range
    percentage_out_of_range = cnt / len(y_pred)
    penalty = 9 * percentage_out_of_range

    print(f"MSE: {mse}, penalty: {penalty}")
    return mse + penalty

# Create an Optuna study and optimize the objective function
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=20)

# Print the best hyperparameters and score
print("Best hyperparameters:", study.best_params)
print("Best validation score:", study.best_value)

# Use the best hyperparameters to train the final model on the full training set
best_params = study.best_params
X_train_full = np.vstack((X_train, X_val))
y_train_full = np.concatenate((y_train, y_val))

best_fm = als.FMRegression(
    n_iter=best_params['n_iter'],
    init_stdev=best_params['init_stdev'],
    rank=best_params['rank'],
    l2_reg_w=best_params['l2_reg_w'],
    l2_reg_V=best_params['l2_reg_V']
)
```

```python
best_fm.fit(X_train, y_train)
```

```python
# Evaluate the final model on the held-out test set
final_predictions = best_fm.predict(X_test)

final_out_of_range = sum([1 for pred in final_predictions if pred < 1 or pred > 5])

# For every prediction that is out of range, assign mean of train prediction
train_mean_rating = sum([i for i in y_train]) / len([i for i in y_train])
for i in range(len(final_predictions)):
    if final_predictions[i] < 1 or final_predictions[i] > 5:
        final_predictions[i] = train_mean_rating
```

```python
import numpy as np

# Calculate RMSE, MSE, and MAE
rmse = np.sqrt(np.mean((final_predictions - y_test) ** 2))
mse = np.mean((final_predictions - y_test) ** 2)
mae = np.mean(np.abs(final_predictions - y_test))

# Print the results
print(f"RMSE: {rmse:.4f}")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
```
